### Assignment 1 (2 scores):

```python
import numpy as np
# Set the random seed for reproducibility
np.random.seed(0)

# Define the covariance matrix
Sigma = np.array([[1, 0], [0, 1]])

# Generate data points
data1 = np.random.multivariate_normal([2, 2], Sigma, 200)
data2 = np.random.multivariate_normal([8, 3], Sigma, 200)
data3 = np.random.multivariate_normal([3, 6], Sigma, 200)

# Combine the datasets
data = np.vstack((data1, data2, data3))
```

```python
def kmeans(data, k, max_iters=100):
    # Randomly initialize centroids
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]

    for _ in range(max_iters):
        # Compute distances from data points to centroids
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        # Assign clusters based on closest centroid
        clusters = np.argmin(distances, axis=1)

        # Update centroids
        new_centroids = np.array([data[clusters == i].mean(axis=0) for i in range(k)])

        # Check for convergence (if centroids do not change)
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return clusters, centroids

# Run K-means
k = 3
clusters, centroids = kmeans(data, k)
```

```python
# Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', marker='o', label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-means Clustering for Assignment 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```
